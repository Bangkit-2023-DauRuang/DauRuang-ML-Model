```python
import os
from bing_image_downloader import downloader
from PIL import Image, ImageStat
```

```python
!mkdir images
```

```python
list = ["carton", "can", "cooking oil", "organic food trash"]

for i in list:
    downloader.download(i,limit=150,output_dir="./images/")
```

```python
#Duplicate checker

img_path = "./images"

for i in list:
    img_item = os.path.join(img_path,i)
    img_file = []
    for i in os.listdir(img_item):
        if i.endswith("jpg") or i.endswith("jpeg"):
            img_file.append(i)
    dup_file = []
    
    for file_org in img_file:
        if not file_org in dup_file:
            img_org = Image.open(os.path.join(img_item, file_org))
            pix_mean1 = ImageStat.Stat(img_org).mean

            for file_check in img_file:
                if file_check != file_org:
                    img_check = Image.open(os.path.join(img_item,file_check))
                    pix_mean2 = ImageStat.Stat(img_check).mean

                    if pix_mean1 == pix_mean2:
                        dup_file.append(file_check)
                        #Removing Original file    
                        os.remove(os.path.join(img_item,file_check))

print(dup_file)
```
